```python
from google.colab import drive
import pandas as pd
import numpy as np
from pprint import pprint as pp

drive.mount('/content/drive')
folder_path = "/content/drive/MyDrive/CMSC723-Final-Project/results/"

combination_names = [

    'baseline-3B-buyer+baseline-3B-seller',
    'baseline-3B-buyer+baseline-COT-3B-seller',

    'baseline-3B-buyer+baseline-8B-seller',
    'baseline-3B-buyer+baseline-COT-8B-seller',

    'baseline-3B-buyer+baseline-70B-seller',
    'baseline-3B-buyer+baseline-COT-70B-seller',

    # ------------------------------------------------

    'baseline-8B-buyer+baseline-3B-seller',
    'baseline-8B-buyer+baseline-COT-3B-seller',

    'baseline-8B-buyer+baseline-8B-seller',
    'baseline-8B-buyer+baseline-COT-8B-seller',

    'baseline-8B-buyer+baseline-70B-seller',
    'baseline-8B-buyer+baseline-COT-70B-seller',

    # ------------------------------------------------

    'baseline-70B-buyer+baseline-3B-seller',
    'baseline-70B-buyer+baseline-COT-3B-seller',

    'baseline-70B-buyer+baseline-8B-seller',
    'baseline-70B-buyer+baseline-COT-8B-seller',

    'baseline-70B-buyer+baseline-70B-seller',
    'baseline-70B-buyer+baseline-COT-70B-seller',

    # ------------------------------------------------

    'baseline-COT-3B-buyer+baseline-3B-seller',
    'baseline-COT-3B-buyer+baseline-COT-3B-seller',

    'baseline-COT-3B-buyer+baseline-8B-seller',
    'baseline-COT-3B-buyer+baseline-COT-8B-seller',

    'baseline-COT-3B-buyer+baseline-70B-seller',
    'baseline-COT-3B-buyer+baseline-COT-70B-seller',

    # ------------------------------------------------

    'baseline-COT-8B-buyer+baseline-3B-seller',
    'baseline-COT-8B-buyer+baseline-COT-3B-seller',

    'baseline-COT-8B-buyer+baseline-8B-seller',
    'baseline-COT-8B-buyer+baseline-COT-8B-seller',

    'baseline-COT-8B-buyer+baseline-70B-seller',
    'baseline-COT-8B-buyer+baseline-COT-70B-seller',

    # ------------------------------------------------

    'baseline-COT-70B-buyer+baseline-3B-seller',
    'baseline-COT-70B-buyer+baseline-COT-3B-seller',

    'baseline-COT-70B-buyer+baseline-8B-seller',
    'baseline-70B-buyer+baseline-COT-8B-seller',

    'baseline-COT-70B-buyer+baseline-70B-seller',
    'baseline-COT-70B-buyer+baseline-COT-70B-seller',
]

combination_dfs = {
    combination_name: pd.read_csv(folder_path + combination_name + '.csv')
    for combination_name in combination_names
}

# Function to calculate concession rate for a single row
def calculate_concession_rate(prices):
    try:
        # Clean invalid prices and ensure it's numeric
        offer_prices = [p for p in eval(prices) if p != -1]
        if len(offer_prices) > 1:
            return sum(abs(b - a) for a, b in zip(offer_prices, offer_prices[1:])) / len(offer_prices)
        else:
            return np.nan
    except:
        return np.nan

# Recalculate concession rates and update DataFrames
for name, df in combination_dfs.items():
    # Apply the concession rate function to each row
    df['concession_rate'] = df['prices'].apply(calculate_concession_rate)
    df.to_csv(folder_path + name + '.csv', index=False)  # Save the updated DataFrame back to CSV

def calculate_relative_eff(fairness, dialogue_length):
    try:
        return abs(fairness) / dialogue_length if dialogue_length > 0 else np.nan
    except Exception:
        return np.nan

# Update relative efficiency in DataFrames
for name, df in combination_dfs.items():
    # Recalculate relative efficiency
    df['rel_efficiency'] = df.apply(lambda row: calculate_relative_eff(row['fairness'], row['dialogue_length']), axis=1)

    # Save updated DataFrame
    df.to_csv(folder_path + name + '.csv', index=False)

aggregated_data = {}
for name, df in combination_dfs.items():
    aggregated_data[name] = {
      'agreement_rate': (df['outcome'] == 'Accepted').sum() / len(df),
      'avg_dialogue_length': df['dialogue_length'].mean(),
      'avg_bias': df['bias'].mean(),
      'avg_aggressiveness': df['aggressiveness'].mean(),
      'avg_fairness': df['fairness'].mean(),
      'avg_bias_cond_dialogue_length': df['bias_cond_dialogue_length'].mean(),
      'avg_concession_rate': df['concession_rate'].mean(),
      'avg_probing_ratio': df['probing_ratio'].mean(),
      'avg_rel_eff': df['rel_efficiency'].mean(),
    }
pp(aggregated_data)
```

```python
"""
TODO
> Plots
  - Across all individual scenarios
    * Bias (-1,0,+1 range) (bar)
    * Bias conditioned on dialogue length (bar)
    * Dialogue Length (bar)
    * Aggressiveness (bar)
    * Fairness (bar)
    * Price progression (line) -> single line plot
  - Combination Plots
    * Average Bias (bar)
    * Average Dialogue Length (bar)
    * Average Bias conditioned on dialogue length (bar)
    * Average Aggressiveness (bar)
    * Average Fairness (bar)
    * Agreement Rate (# agreements / # num_transcripts) (bar)
  - Role Aggregated Plots
    * Average Bias (bar)
    * Average Dialogue Length (bar)
    * Average Bias conditioned on dialogue length (bar)
    * Average Aggressiveness
    * Average Fairness (bar)
    * Aggreement rate (bar)
    * Listing price vs aggressiveness (scatter)
    * Listing price vs bias (scatter)
    * Listing price vs dialogue length (scatter)
  - Inspection
    * Scenario information
    * Display negotiation
    * Price progression (line)
> Widgets
"""

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import ast
import numpy as np
import random

def random_palette(n_colors):
    palette_options = ["husl", "coolwarm", "viridis", "cubehelix", "muted", "Set2", "Paired"]
    selected_palette = random.choice(palette_options)
    return sns.color_palette(selected_palette, n_colors)

def process_prices(prices, listing_price):
      updated_prices = []
      recent_price = listing_price
      for price in prices:
        if price == -1:
          updated_prices.append(recent_price)
        else:
          updated_prices.append(price)
          recent_price = price
      return updated_prices

# a combination_df is just the result_df from a buyer-seller combination like:
# baseline-Llama-3.2-3B-buyer+baseline-Llama-3.2-3B-seller.csv
class ScenarioPlots:
  def __init__(self, combination_name):
    self.combination_df = pd.read_csv(folder_path + combination_name + '.csv')
    self.combination_df['prices'] = self.combination_df['prices'].apply(
        lambda x: ast.literal_eval(x)
    )
    self.combination_df['prices'] = self.combination_df.apply(
        lambda x: process_prices(x['prices'], x['listing_price']),
        axis=1
    )

  def show_bar_plot(self, y_col_name, plot_title, x_label, y_label):
    self.combination_df['abbrev_title'] = [f"S{i+1}" for i in range(len(self.combination_df))]

    plt.figure(figsize=(10,6))
    sns.barplot(
        x='abbrev_title',
        y=y_col_name,
        data=self.combination_df,
        hue='abbrev_title',
        legend=False,
        palette=random_palette(len(self.combination_df))
    )
    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

  def show_bias_bar_plot(self, bias_type, plot_title, x_label, y_label):
    self.combination_df['abv_title'] = [f"S{i+1}" for i in range(len(self.combination_df))]
    self.combination_df['color_cat'] = self.combination_df[bias_type].apply(lambda x: 'Biased Towards Buyer' if x > 0 else 'Biased Towards Seller')
    color_palette = {'Biased Towards Buyer': 'green', 'Biased Towards Seller': 'red'}

    plt.figure(figsize=(10,6))
    sns.barplot(x='abv_title', y=bias_type, data=self.combination_df, palette=color_palette, hue='color_cat')
    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)

    plt.legend(title="Bias Legend", loc='upper right', bbox_to_anchor=(1.30, 1))
    plt.show()

  def show_muti_line_plot(self, plot_title, x_label, y_label):
    df_expanded = self.combination_df.explode('prices')
    df_expanded['Turn'] = df_expanded.groupby('title').cumcount() + 1
    df_expanded['prices'] = pd.to_numeric(df_expanded['prices'])

    plt.figure(figsize=(10,6))
    sns.lineplot(
        data=df_expanded,
        x='Turn',
        y='prices',
        hue='title',
        marker = 'o',
    )
    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend(title='title', bbox_to_anchor=(1, 1))
    plt.show()

  def show_listing_price_comp_line_plot(self, comp_cat, plot_title, x_label, y_label):
    grouped_df = self.combination_df.groupby('listing_price')[comp_cat].mean().reset_index()

    grouped_df = grouped_df.dropna(subset=[comp_cat])

    x = grouped_df['listing_price']
    y = grouped_df[comp_cat]


    m, b = np.polyfit(x, y, 1)
    bfl = m * x + b

    plt.figure(figsize=(10,6))
    plt.plot(x, bfl, color='pink', linestyle ='-', label='best fit line')
    plt.scatter(x, y, label='comp_cat')

    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()


# class CombinationPlots:
#   def __init__(self, combination_names):
#     self.combination_dfs = {
#         combination_name: pd.read_csv(folder_path + combination_name + '.csv')
#         for combination_name in combination_names
#     }
#   def show_combination_bar_plot(self, y_col_name, plot_title, x_label, y_label):
#     aggregated_data = {}
#     for name, df in self.combination_dfs.items():
#         aggregated_data[name] = {
#           'agreement_rate': (df['outcome'] == 'Accepted').sum() / len(df),
#           'avg_dialogue_length': df['dialogue_length'].mean(),
#           'avg_bias': df['bias'].mean(),
#           'avg_aggressiveness': df['aggressiveness'].mean(),
#           'avg_fairness': df['fairness'].mean(),
#           'avg_bias_cond_dialogue_length': df['bias_cond_dialogue_length'].mean(),
#       }

#     data = {
#         name: metrics[y_col_name] for name, metrics in aggregated_data.items()
#     }

#     df_plot = pd.DataFrame(list(data.items()), columns=['Combination', y_col_name])
#     plt.figure(figsize=(10,6))
#     sns.barplot(
#         x='Combination',
#         y=y_col_name,
#         data=df_plot,
#         hue='Combination',
#         legend=False,
#         palette=random_palette(len(df_plot))
#     )

#     plt.xticks(rotation=80)

#     plt.title(plot_title)
#     plt.xlabel(x_label)
#     plt.ylabel(y_label)
#     plt.show()

class CombinationPlots:
    def __init__(self, combination_names):
        self.combination_dfs = {
            combination_name: pd.read_csv(folder_path + combination_name + '.csv')
            for combination_name in combination_names
        }

    def parse_combination_name(self, combo_name):
        buyer_part, seller_part = combo_name.split('+')
        buyer_parts = buyer_part.split('-')
        seller_parts = seller_part.split('-')

        # Determine Buyer Type and Size
        if 'COT' in buyer_parts:
            buyer_type = 'COT'
            buyer_size = buyer_parts[2] if buyer_parts.index('COT') == 1 else buyer_parts[1]
        else:
            buyer_type = 'base'
            buyer_size = buyer_parts[1]

        # Determine Seller Type and Size
        if 'COT' in seller_parts:
            seller_type = 'COT'
            seller_size = seller_parts[2] if seller_parts.index('COT') == 1 else seller_parts[1]
        else:
            seller_type = 'base'
            seller_size = seller_parts[1]

        return buyer_type, buyer_size, seller_type, seller_size

    def show_combination_bar_plot(self, y_col_name, plot_title, x_label, y_label):
        # Calculate aggregated metrics
        aggregated_data = {}
        for name, df in self.combination_dfs.items():
            aggregated_data[name] = {
              'agreement_rate': (df['outcome'] == 'Accepted').sum() / len(df),
              'avg_dialogue_length': df['dialogue_length'].mean(),
              'avg_bias': df['bias'].mean(),
              'avg_aggressiveness': df['aggressiveness'].mean(),
              'avg_fairness': df['fairness'].mean(),
              'avg_bias_cond_dialogue_length': df['bias_cond_dialogue_length'].mean(),
              'avg_concession_rate': df['concession_rate'].mean(),
              'avg_probing_ratio': df['probing_ratio'].mean(),
              'avg_concession_rate': df['concession_rate'].mean(),
              'avg_probing_ratio': df['probing_ratio'].mean(),
              'avg_rel_eff': df['rel_efficiency'].mean(),
            }

        # Create DataFrame from aggregated data
        data = {name: metrics[y_col_name] for name, metrics in aggregated_data.items()}
        df_plot = pd.DataFrame(list(data.items()), columns=['Combination', y_col_name])

        # Parse combination names into structured columns
        df_plot[['BuyerType', 'BuyerSize', 'SellerType', 'SellerSize']] = df_plot.apply(
            lambda row: self.parse_combination_name(row['Combination']), axis=1, result_type='expand'
        )

        # Define orders
        size_order = ['3B', '8B', '70B']
        type_order = ['base', 'COT']

        df_plot['BuyerSize'] = pd.Categorical(df_plot['BuyerSize'], categories=size_order, ordered=True)
        df_plot['SellerSize'] = pd.Categorical(df_plot['SellerSize'], categories=size_order, ordered=True)
        df_plot['BuyerType'] = pd.Categorical(df_plot['BuyerType'], categories=type_order, ordered=True)
        df_plot['SellerType'] = pd.Categorical(df_plot['SellerType'], categories=type_order, ordered=True)

        # Determine if combination has COT
        df_plot['HasCOT'] = df_plot.apply(lambda x: x['BuyerType'] == 'COT' or x['SellerType'] == 'COT', axis=1)

        # Split into two groups: No-COT and COT
        df_no_cot = df_plot[df_plot['HasCOT'] == False].copy()
        df_cot = df_plot[df_plot['HasCOT'] == True].copy()

        # Sort each group
        df_no_cot = df_no_cot.sort_values(by=['BuyerSize', 'SellerSize', 'BuyerType', 'SellerType'])
        df_cot = df_cot.sort_values(by=['BuyerSize', 'SellerSize', 'BuyerType', 'SellerType'])

        # Insert multiple gap rows between no_cot and cot groups for a large space
        gap_count = 3  # number of gap bars for a larger space
        gap_rows = []
        for i in range(gap_count):
            gap_rows.append({
                'Combination': f'(gap{i})',
                y_col_name: 0,
                'BuyerSize': size_order[0],
                'BuyerType': type_order[0],
                'SellerSize': size_order[0],
                'SellerType': type_order[0],
                'HasCOT': False
            })
        gap_df = pd.DataFrame(gap_rows)

        # Concatenate: No-COT group + gap + COT group
        df_final = pd.concat([df_no_cot, gap_df, df_cot], ignore_index=True)

        # Create DisplayName
        df_final['DisplayName'] = df_final.apply(
            lambda x: '' if x['Combination'].startswith('(gap') else f"{x['BuyerSize']}-B({x['BuyerType']})+{x['SellerSize']}-S({x['SellerType']})",
            axis=1
        )

        # Determine order from final DataFrame
        order = df_final['DisplayName'].unique().tolist()

        # Identify normal (non-gap) categories for no-cot and cot separately
        no_cot_names = df_no_cot.apply(lambda x: f"{x['BuyerSize']}-B({x['BuyerType']})+{x['SellerSize']}-S({x['SellerType']})", axis=1).unique().tolist()
        cot_names = df_cot.apply(lambda x: f"{x['BuyerSize']}-B({x['BuyerType']})+{x['SellerSize']}-S({x['SellerType']})", axis=1).unique().tolist()
        gap_names = [n for n in order if n == '']

        # Assign palettes: One for no-cot, one for cot
        pal_no_cot = random_palette(len(no_cot_names))
        pal_cot = random_palette(len(cot_names))

        pal_dict = {}
        # Assign no-cot colors
        for name, color in zip(no_cot_names, pal_no_cot):
            pal_dict[name] = color

        # Assign cot colors
        for name, color in zip(cot_names, pal_cot):
            pal_dict[name] = color

        # Gaps to white
        for g in gap_names:
            pal_dict[g] = (1.0, 1.0, 1.0)  # white

        plt.figure(figsize=(14, 6))
        sns.barplot(
            x='DisplayName',
            y=y_col_name,
            data=df_final,
            order=order,
            palette=pal_dict,
            ci=None
        )
        plt.xticks(rotation=45, ha='right')
        plt.title(plot_title)
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.tight_layout()
        plt.show()



# TODO
class RoleAggregatedPlots:
  def __init__(self):
    pass

class InspectScenario:
  def __init__(self, combination_name):
    self.combination_df = pd.read_csv(folder_path + combination_name + '.csv')
    self.scenario_names = [
        f"Scenario {i+1}: " + title
        for i,title in enumerate(self.combination_df['title'].tolist())
    ]
    self.combination_df['prices'] = self.combination_df['prices'].apply(
        lambda x: ast.literal_eval(x)
    )

    self.combination_df['prices'] = self.combination_df.apply(
        lambda x: process_prices(x['prices'], x['listing_price']),
        axis=1
    )

  def visualize_scenario(self, name):

    scenario_title = name.split(": ")[1]
    scenario_data = self.combination_df[self.combination_df['title'] == scenario_title]
    prices = scenario_data['prices'].iloc[0]
    turns = range(1, len(prices) + 1)

    s = scenario_data.iloc[0]

    print("\n--- Scenario Info ---\n")
    print(f"Title:                  {s['title']}")
    print(f"Description:            {s['description']}")
    print(f"Category:               {s['category']}")
    print(f"Listing Price:          ${s['listing_price']}")
    print(f"Buyer Target:           ${s['buyer_target']}")
    print(f"Seller Target:          ${s['seller_target']}")

    print("\n--- Metrics ---\n")
    print(f"Bias:                   {s['bias']}")
    print(f"Aggressiveness:         {s['aggressiveness']}")
    print(f"Fairness:               {s['fairness']}")
    print(f"Dialogue Length:        {s['dialogue_length']}")
    print(f"Bias::Dialogue Length:  {s['bias_cond_dialogue_length']}")

    print("\n--- Transcript ---")
    print(s['transcript_display'])

    print("\n--- Price Progression Plot ---\n")
    plt.figure(figsize=(10,6))
    plt.plot(turns, prices, marker='o', label=scenario_title)
    plt.title(f'Price Progression for <{scenario_title}>')
    plt.xlabel('Turn')
    plt.ylabel('Price')
    plt.show()
```

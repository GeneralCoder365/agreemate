```python
from google.colab import drive
import pandas as pd
import numpy as np
from pprint import pprint as pp

drive.mount('/content/drive')
folder_path = "/content/drive/MyDrive/CMSC723-Final-Project/results/"

combination_names = [

    'baseline-3B-buyer+baseline-3B-seller',
    'baseline-3B-buyer+baseline-COT-3B-seller',

    'baseline-3B-buyer+baseline-8B-seller',
    'baseline-3B-buyer+baseline-COT-8B-seller',

    'baseline-3B-buyer+baseline-70B-seller',
    'baseline-3B-buyer+baseline-COT-70B-seller',

    # ------------------------------------------------

    'baseline-8B-buyer+baseline-3B-seller',
    'baseline-8B-buyer+baseline-COT-3B-seller',

    'baseline-8B-buyer+baseline-8B-seller',
    'baseline-8B-buyer+baseline-COT-8B-seller',

    'baseline-8B-buyer+baseline-70B-seller',
    'baseline-8B-buyer+baseline-COT-70B-seller',

    # ------------------------------------------------

    'baseline-70B-buyer+baseline-3B-seller',
    'baseline-70B-buyer+baseline-COT-3B-seller',

    'baseline-70B-buyer+baseline-8B-seller',
    'baseline-70B-buyer+baseline-COT-8B-seller',

    'baseline-70B-buyer+baseline-70B-seller',
    'baseline-70B-buyer+baseline-COT-70B-seller',

    # ------------------------------------------------

    'baseline-COT-3B-buyer+baseline-3B-seller',
    'baseline-COT-3B-buyer+baseline-COT-3B-seller',

    'baseline-COT-3B-buyer+baseline-8B-seller',
    'baseline-COT-3B-buyer+baseline-COT-8B-seller',

    'baseline-COT-3B-buyer+baseline-70B-seller',
    'baseline-COT-3B-buyer+baseline-COT-70B-seller',

    # ------------------------------------------------

    'baseline-COT-8B-buyer+baseline-3B-seller',
    'baseline-COT-8B-buyer+baseline-COT-3B-seller',

    'baseline-COT-8B-buyer+baseline-8B-seller',
    'baseline-COT-8B-buyer+baseline-COT-8B-seller',

    'baseline-COT-8B-buyer+baseline-70B-seller',
    'baseline-COT-8B-buyer+baseline-COT-70B-seller',

    # ------------------------------------------------

    'baseline-COT-70B-buyer+baseline-3B-seller',
    'baseline-COT-70B-buyer+baseline-COT-3B-seller',

    'baseline-COT-70B-buyer+baseline-8B-seller',
    'baseline-70B-buyer+baseline-COT-8B-seller',

    'baseline-COT-70B-buyer+baseline-70B-seller',
    'baseline-COT-70B-buyer+baseline-COT-70B-seller',
]

combination_dfs = {
    combination_name: pd.read_csv(folder_path + combination_name + '.csv')
    for combination_name in combination_names
}

# Function to calculate concession rate for a single row
def calculate_concession_rate(prices):
    try:
        # Clean invalid prices and ensure it's numeric
        offer_prices = [p for p in eval(prices) if p != -1]
        if len(offer_prices) > 1:
            return sum(abs(b - a) for a, b in zip(offer_prices, offer_prices[1:])) / len(offer_prices)
        else:
            return np.nan
    except:
        return np.nan

for name, df in combination_dfs.items():
    df['concession_rate'] = df['prices'].apply(calculate_concession_rate)
    df.to_csv(folder_path + name + '.csv', index=False)

def calculate_relative_eff(fairness, dialogue_length):
    try:
        return abs(fairness) / dialogue_length if dialogue_length > 0 else np.nan
    except Exception:
        return np.nan

for name, df in combination_dfs.items():
    df['rel_efficiency'] = df.apply(lambda row: calculate_relative_eff(row['fairness'], row['dialogue_length']), axis=1)
    df.to_csv(folder_path + name + '.csv', index=False)

aggregated_data = {}
for name, df in combination_dfs.items():
    aggregated_data[name] = {
      'agreement_rate': (df['outcome'] == 'Accepted').sum() / len(df),
      'avg_dialogue_length': df['dialogue_length'].mean(),
      'avg_bias': df['bias'].mean(),
      'avg_aggressiveness': df['aggressiveness'].mean(),
      'avg_fairness': df['fairness'].mean(),
      'avg_bias_cond_dialogue_length': df['bias_cond_dialogue_length'].mean(),
      'avg_concession_rate': df['concession_rate'].mean(),
      'avg_probing_ratio': df['probing_ratio'].mean(),
      'avg_rel_eff': df['rel_efficiency'].mean(),
    }
pp(aggregated_data)
```

```python
"""
TODO
> Plots
  - Across all individual scenarios
    * Bias (-1,0,+1 range) (bar)
    * Bias conditioned on dialogue length (bar)
    * Dialogue Length (bar)
    * Aggressiveness (bar)
    * Fairness (bar)
    * Price progression (line) -> single line plot
  - Combination Plots
    * Average Bias (bar)
    * Average Dialogue Length (bar)
    * Average Bias conditioned on dialogue length (bar)
    * Average Aggressiveness (bar)
    * Average Fairness (bar)
    * Agreement Rate (# agreements / # num_transcripts) (bar)
  - Role Aggregated Plots
    * Average Bias (bar)
    * Average Dialogue Length (bar)
    * Average Bias conditioned on dialogue length (bar)
    * Average Aggressiveness
    * Average Fairness (bar)
    * Aggreement rate (bar)
    * Listing price vs aggressiveness (scatter)
    * Listing price vs bias (scatter)
    * Listing price vs dialogue length (scatter)
  - Inspection
    * Scenario information
    * Display negotiation
    * Price progression (line)
> Widgets
"""

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import ast
import numpy as np
import random

def random_palette(n_colors):
    palette_options = ["husl", "coolwarm", "viridis", "cubehelix", "muted", "Set2", "Paired"]
    selected_palette = random.choice(palette_options)
    return sns.color_palette(selected_palette, n_colors)

def process_prices(prices, listing_price):
      updated_prices = []
      recent_price = listing_price
      for price in prices:
        if price == -1:
          updated_prices.append(recent_price)
        else:
          updated_prices.append(price)
          recent_price = price
      return updated_prices

# a combination_df is just the result_df from a buyer-seller combination like:
# baseline-Llama-3.2-3B-buyer+baseline-Llama-3.2-3B-seller.csv
class ScenarioPlots:
  def __init__(self, combination_name):
    self.combination_df = pd.read_csv(folder_path + combination_name + '.csv')
    self.combination_df['prices'] = self.combination_df['prices'].apply(
        lambda x: ast.literal_eval(x)
    )
    self.combination_df['prices'] = self.combination_df.apply(
        lambda x: process_prices(x['prices'], x['listing_price']),
        axis=1
    )

  def show_bar_plot(self, y_col_name, plot_title, x_label, y_label):
    self.combination_df['abbrev_title'] = [f"S{i+1}" for i in range(len(self.combination_df))]

    plt.figure(figsize=(10,6))
    sns.barplot(
        x='abbrev_title',
        y=y_col_name,
        data=self.combination_df,
        hue='abbrev_title',
        legend=False,
        palette=random_palette(len(self.combination_df))
    )
    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

  def show_bias_bar_plot(self, bias_type, plot_title, x_label, y_label):
    self.combination_df['abv_title'] = [f"S{i+1}" for i in range(len(self.combination_df))]
    self.combination_df['color_cat'] = self.combination_df[bias_type].apply(lambda x: 'Biased Towards Buyer' if x > 0 else 'Biased Towards Seller')
    color_palette = {'Biased Towards Buyer': 'green', 'Biased Towards Seller': 'red'}

    plt.figure(figsize=(10,6))
    sns.barplot(x='abv_title', y=bias_type, data=self.combination_df, palette=color_palette, hue='color_cat')
    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)

    plt.legend(title="Bias Legend", loc='upper right', bbox_to_anchor=(1.30, 1))
    plt.show()

  def show_muti_line_plot(self, plot_title, x_label, y_label):
    df_expanded = self.combination_df.explode('prices')
    df_expanded['Turn'] = df_expanded.groupby('title').cumcount() + 1
    df_expanded['prices'] = pd.to_numeric(df_expanded['prices'])

    plt.figure(figsize=(10,6))
    sns.lineplot(
        data=df_expanded,
        x='Turn',
        y='prices',
        hue='title',
        marker = 'o',
    )
    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend(title='title', bbox_to_anchor=(1, 1))
    plt.show()

  def show_listing_price_comp_line_plot(self, comp_cat, plot_title, x_label, y_label):
    grouped_df = self.combination_df.groupby('listing_price')[comp_cat].mean().reset_index()

    grouped_df = grouped_df.dropna(subset=[comp_cat])

    x = grouped_df['listing_price']
    y = grouped_df[comp_cat]


    m, b = np.polyfit(x, y, 1)
    bfl = m * x + b

    plt.figure(figsize=(10,6))
    plt.plot(x, bfl, color='pink', linestyle ='-', label='best fit line')
    plt.scatter(x, y, label='comp_cat')

    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()


class CombinationPlots:
  def __init__(self, combination_names):
    self.combination_dfs = {
        combination_name: pd.read_csv(folder_path + combination_name + '.csv')
        for combination_name in combination_names
    }
  def show_combination_bar_plot(self, y_col_name, plot_title, x_label, y_label):
    aggregated_data = {}
    for name, df in self.combination_dfs.items():
        aggregated_data[name] = {
          'agreement_rate': (df['outcome'] == 'Accepted').sum() / len(df),
          'avg_dialogue_length': df['dialogue_length'].mean(),
          'avg_bias': df['bias'].mean(),
          'avg_aggressiveness': df['aggressiveness'].mean(),
          'avg_fairness': df['fairness'].mean(),
          'avg_bias_cond_dialogue_length': df['bias_cond_dialogue_length'].mean(),
      }

    data = {
        name: metrics[y_col_name] for name, metrics in aggregated_data.items()
    }

    df_plot = pd.DataFrame(list(data.items()), columns=['Combination', y_col_name])
    plt.figure(figsize=(10,6))
    sns.barplot(
        x='Combination',
        y=y_col_name,
        data=df_plot,
        hue='Combination',
        legend=False,
        palette=random_palette(len(df_plot))
    )

    plt.xticks(rotation=80)

    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()



# TODO
class RoleAggregatedPlots:
  def __init__(self):
    pass

class InspectScenario:
  def __init__(self, combination_name):
    self.combination_df = pd.read_csv(folder_path + combination_name + '.csv')
    self.scenario_names = [
        f"Scenario {i+1}: " + title
        for i,title in enumerate(self.combination_df['title'].tolist())
    ]
    self.combination_df['prices'] = self.combination_df['prices'].apply(
        lambda x: ast.literal_eval(x)
    )

    self.combination_df['prices'] = self.combination_df.apply(
        lambda x: process_prices(x['prices'], x['listing_price']),
        axis=1
    )

  def visualize_scenario(self, name):

    scenario_title = name.split(": ")[1]
    scenario_data = self.combination_df[self.combination_df['title'] == scenario_title]
    prices = scenario_data['prices'].iloc[0]
    turns = range(1, len(prices) + 1)

    s = scenario_data.iloc[0]

    print("\n--- Scenario Info ---\n")
    print(f"Title:                  {s['title']}")
    print(f"Description:            {s['description']}")
    print(f"Category:               {s['category']}")
    print(f"Listing Price:          ${s['listing_price']}")
    print(f"Buyer Target:           ${s['buyer_target']}")
    print(f"Seller Target:          ${s['seller_target']}")

    print("\n--- Metrics ---\n")
    print(f"Bias:                   {s['bias']}")
    print(f"Aggressiveness:         {s['aggressiveness']}")
    print(f"Fairness:               {s['fairness']}")
    print(f"Dialogue Length:        {s['dialogue_length']}")
    print(f"Bias::Dialogue Length:  {s['bias_cond_dialogue_length']}")

    print("\n--- Transcript ---")
    print(s['transcript_display'])

    print("\n--- Price Progression Plot ---\n")
    plt.figure(figsize=(10,6))
    plt.plot(turns, prices, marker='o', label=scenario_title)
    plt.title(f'Price Progression for <{scenario_title}>')
    plt.xlabel('Turn')
    plt.ylabel('Price')
    plt.show()
```

```python
import ipywidgets as widgets
from IPython.display import display

# 'finetune-3B-buyer+finetune-3B-seller',

# add new combination names here
combination_names = [
    'baseline-COT-8B-buyer+baseline-COT-8B-seller',
    'baseline-COT-70B-buyer+baseline-COT-70B-seller',
    'baseline-3B-buyer-aggressive+baseline-3B-seller-aggressive',
    'baseline-3B-buyer-aggressive+baseline-3B-seller-passive',
]

mode_radio = widgets.RadioButtons(
    options=[
        'Scenario',
        'Combination',
        'Role Aggregated',
        'Inspection'
    ],
    value='Scenario',
    layout={'width': 'max-content'},
    description='Mode:',
    disabled=False,
    style={'description_width': '150px'}
)

output = widgets.Output()
mode_container = widgets.VBox([mode_radio, output])

def on_value_change(change):
    mode_container.children = []
    with output:
        output.clear_output()
        if change['new'] == 'Combination':
            combination_plots_obj = CombinationPlots(combination_names)
            combination_plot_dropdown = widgets.Dropdown(
                options=[
                    'Agreement Rate',
                    'Average Bias',
                    'Average Bias Conditioned On Dialogue Length',
                    'Average Dialogue Length',
                    'Average Aggressiveness',
                    'Average Fairness',
                    'Average Probing',
                    'Average Concession Rate',
                    'Average Relative Efficiency',
                ],
                value='Agreement Rate',
                description='Plot:',
                disabled=False,
                style={'description_width': '150px'},
                layout={'width': 'max-content'}
            )
            display(combination_plot_dropdown)
            combination_plot_dropdown_output = widgets.Output()

            combo_visualize_button = widgets.Button(description="Visualize", layout={'margin': '0 0 0 160px'})
            combo_visualize_button_output = widgets.Output()
            display(combo_visualize_button)

            def combo_visualize_button_click(b):
                combo_visualize_button_output.clear_output()
                with combo_visualize_button_output:
                    if combination_plot_dropdown.value == 'Agreement Rate':
                        combination_plots_obj.show_combination_bar_plot(
                            'agreement_rate',
                            'Agreement Rate Across Combinations',
                            'Combination',
                            'Agreement Rate'
                        )
                    elif combination_plot_dropdown.value == 'Average Bias':
                        combination_plots_obj.show_combination_bar_plot(
                            'avg_bias',
                            'Average Bias Across Combinations',
                            'Combination',
                            'Average Bias'
                        )
                    elif combination_plot_dropdown.value == \
                        'Average Bias Conditioned On Dialogue Length':
                        combination_plots_obj.show_combination_bar_plot(
                            'avg_bias_cond_dialogue_length',
                            'Average Bias Conditioned On Dialogue Length Across Combinations',
                            'Combination',
                            'Average Bias'
                        )
                    elif combination_plot_dropdown.value == 'Average Dialogue Length':
                        combination_plots_obj.show_combination_bar_plot(
                            'avg_dialogue_length',
                            'Average Dialogue Length Across Combinations',
                            'Combination',
                            'Average Dialogue Length'
                        )
                    elif combination_plot_dropdown.value == 'Average Aggressiveness':
                        combination_plots_obj.show_combination_bar_plot(
                            'avg_aggressiveness',
                            'Average Aggressiveness Across Combinations',
                            'Combination',
                            'Average Aggressiveness'
                        )
                    elif combination_plot_dropdown.value == 'Average Fairness':
                        combination_plots_obj.show_combination_bar_plot(
                            'avg_fairness',
                            'Average Fairness Across Combinations',
                            'Combination',
                            'Average Fairness'
                        )

                    elif combination_plot_dropdown.value == 'Average Relative Efficiency':
                        combination_plots_obj.show_combination_bar_plot(
                            'avg_rel_eff',
                            'Average Relative Efficiency Across Combinations',
                            'Combination',
                            'Average Relative Efficiency'
                        )

                    elif combination_plot_dropdown.value == 'Average Probing':
                        combination_plots_obj.show_combination_bar_plot(
                            'avg_probing_ratio',
                            'Average Average Probing Across Combinations',
                            'Combination',
                            'Average Average Probing'
                        )
                    elif combination_plot_dropdown.value == 'Average Concession Rate':
                        combination_plots_obj.show_combination_bar_plot(
                            'avg_concession_rate',
                            'Average Average Probing Across Combinations',
                            'Combination',
                            'Average Average Probing'
                        )
            combo_visualize_button.on_click(combo_visualize_button_click)
            display(combo_visualize_button_output)

            mode_container.children = [
                combination_plot_dropdown,
                combo_visualize_button,
                combo_visualize_button_output
            ]

        elif change['new'] == 'Scenario':
            scenario_comb_name_dropdown = widgets.Dropdown(
                options=combination_names,
                value=combination_names[0],
                description='Buyer-Seller Pair:',
                disabled=False,
                style={'description_width': '150px'},
                layout={'width': 'max-content'}
            )
            display(scenario_comb_name_dropdown)
            scenario_comb_name_dropdown_output = widgets.Output()

            scenario_plots_obj = ScenarioPlots(scenario_comb_name_dropdown.value)

            def on_scenario_comb_name_dropdown_change(change):
                nonlocal scenario_plots_obj
                scenario_plots_obj = ScenarioPlots(change['new'])
            scenario_comb_name_dropdown.observe(on_scenario_comb_name_dropdown_change, names='value')

            scenario_plot_dropdown = widgets.Dropdown(
                options=[
                    'Bias Across Scenarios',
                    'Bias Conditioned On Dialogue Length',
                    'Dialogue Length Across Scenarios',
                    'Aggressiveness Across Scenarios',
                    'Fairness Across Scenarios',
                    'Price Progression',
                    'Listing Price vs Aggressiveness',
                    'Listing Price vs Bias',
                    'Listing Price vs Dialogue Length'
                ],
                value='Bias Across Scenarios',
                description='Plot:',
                disabled=False,
                style={'description_width': '150px'},
                layout={'width': 'max-content'}
            )
            display(scenario_plot_dropdown)
            scenario_plot_dropdown_output = widgets.Output()

            visualize_button = widgets.Button(description="Visualize",
                                              layout={'margin': '0 0 0 160px'})
            visualize_button_output = widgets.Output()
            display(visualize_button)

            def on_visualize_button_click(b):
                visualize_button_output.clear_output()
                with visualize_button_output:
                    if scenario_plot_dropdown.value == 'Bias Across Scenarios':
                        scenario_plots_obj.show_bias_bar_plot(
                            'bias',
                            'Bias Across Scenarios',
                            'Scenario',
                            'Bias'
                        )
                    if scenario_plot_dropdown.value == 'Bias Conditioned On Dialogue Length':
                        scenario_plots_obj.show_bias_bar_plot(
                            'bias_cond_dialogue_length',
                            'Bias Conditioned On Dialogue Length Across Scenarios',
                            'Scenario',
                            'Bias'
                        )
                    elif scenario_plot_dropdown.value == 'Dialogue Length Across Scenarios':
                        scenario_plots_obj.show_bar_plot(
                            'dialogue_length',
                            'Dialogue Length Across Scenarios',
                            'Scenario',
                            'Dialogue Length'
                        )
                    elif scenario_plot_dropdown.value == 'Aggressiveness Across Scenarios':
                        scenario_plots_obj.show_bar_plot(
                            'aggressiveness',
                            'Aggressiveness Across Scenarios',
                            'Scenario',
                            'Aggressiveness'
                        )
                    elif scenario_plot_dropdown.value == 'Fairness Across Scenarios':
                        scenario_plots_obj.show_bar_plot(
                            'fairness',
                            'Fairness Across Scenarios',
                            'Scenario',
                            'Fairness'
                        )
                    elif scenario_plot_dropdown.value == 'Price Progression':
                        scenario_plots_obj.show_muti_line_plot(
                            'Price Progression',
                            'Turn',
                            'Price'
                        )
                    elif scenario_plot_dropdown.value == 'Listing Price vs Aggressiveness':
                        scenario_plots_obj.show_listing_price_comp_line_plot(
                            'aggressiveness',
                            'Listing Price vs Aggressiveness',
                            'Listing Price',
                            'Aggressiveness'
                        )
                    elif scenario_plot_dropdown.value == 'Listing Price vs Bias':
                        scenario_plots_obj.show_listing_price_comp_line_plot(
                            'bias',
                            'Listing Price vs Bias',
                            'Listing Price',
                            'Bias'
                        )
                    elif  scenario_plot_dropdown.value == 'Listing Price vs Dialogue Length':
                        scenario_plots_obj.show_listing_price_comp_line_plot(
                            'dialogue_length',
                            'Listing Price vs Dialogue Length',
                            'Listing Price',
                            'Dialogue Length',
                        )
            visualize_button.on_click(on_visualize_button_click)
            display(visualize_button_output)

            mode_container.children = [
                scenario_comb_name_dropdown,
                scenario_plot_dropdown,
                visualize_button,
                visualize_button_output
            ]
        elif change['new'] == 'Inspection':
            inspection_comb_name_dropdown = widgets.Dropdown(
                options=combination_names,
                value=combination_names[0],
                description='Buyer-Seller Pair:',
                disabled=False,
                style={'description_width': '150px'},
                layout={'width': 'max-content'}
            )
            display(inspection_comb_name_dropdown)
            inspection_comb_name_dropdown_output = widgets.Output()

            inspection_plots_obj = InspectScenario(inspection_comb_name_dropdown.value)

            inspection_scenario_name_dropdown = widgets.Dropdown(
                options=inspection_plots_obj.scenario_names,
                value=inspection_plots_obj.scenario_names[0],
                description='Scenario:',
                disabled=False,
                style={'description_width': '150px'},
                layout={'width': 'max-content'}
            )
            display(inspection_scenario_name_dropdown)
            inspection_scenario_name_dropdown_output = widgets.Output()

            visualize_scenario_button = widgets.Button(
                description="Visualize",
                layout={'margin': '0 0 0 160px'}
            )
            display(visualize_scenario_button)
            visualize_scenario_button_output = widgets.Output()

            def on_inspection_comb_name_dropdown_change(change):
                nonlocal inspection_plots_obj
                inspection_plots_obj = InspectScenario(change['new'])
                inspection_scenario_name_dropdown.options = inspection_plots_obj.scenario_names
                inspection_scenario_name_dropdown.value = inspection_plots_obj.scenario_names[0]

            def on_visualize_scenario_button_click(b):
                visualize_scenario_button_output.clear_output()
                with visualize_scenario_button_output:
                    inspection_plots_obj.visualize_scenario(inspection_scenario_name_dropdown.value)

            inspection_comb_name_dropdown.observe(on_inspection_comb_name_dropdown_change, names='value')
            visualize_scenario_button.on_click(on_visualize_scenario_button_click)

            mode_container.children = [
              inspection_comb_name_dropdown,
              inspection_scenario_name_dropdown,
              visualize_scenario_button,
              visualize_scenario_button_output
            ]


mode_radio.observe(on_value_change, names='value')
on_value_change({'new': mode_radio.value})
display(widgets.VBox([mode_radio, mode_container]))
```
